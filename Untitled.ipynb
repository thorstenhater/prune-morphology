# Allen Cell.



```python
from collections import defaultdict
from dataclasses import dataclass
from typing import Optional
import json
from pathlib import Path

def load_allen_fit(fit):
    with open(fit) as fd:
        fit = json.load(fd)

    # cable parameters convenience class
    @dataclass
    class parameters:
        cm: Optional[U.quantity] = None
        temp: Optional[U.quantity] = None
        Vm: Optional[U.quantity] = None
        rL: Optional[U.quantity] = None

    param = defaultdict(parameters)
    mechs = defaultdict(dict)
    for block in fit["genome"]:
        mech = block["mechanism"] or "pas"
        region = block["section"]
        name = block["name"]
        value = float(block["value"])
        if name.endswith("_" + mech):
            name = name[: -(len(mech) + 1)]
        elif mech == "pas":
            # transform names and values
            if name == "cm":
                param[region].cm = value * U.uF / U.cm2
            elif name == "Ra":
                param[region].rL = value * U.Ohm * U.cm
            elif name == "Vm":
                param[region].Vm = value * U.mV
            elif name == "celsius":
                param[region].temp = value * U.Celsius
            else:
                raise Exception(f"Unknown key: {name}")
            continue
        else:
            raise Exception(f"Illegal combination {mech} {name}")
        mechs[(region, mech)][name] = value

    regs = list(param.items())
    mechs = [(r, m, vs) for (r, m), vs in mechs.items()]

    default = parameters(
        temp=float(fit["conditions"][0]["celsius"]) * U.Celsius,
        Vm=float(fit["conditions"][0]["v_init"]) * U.mV,
        rL=float(fit["passive"][0]["ra"]) * U.Ohm * U.cm,
    )

    ions = []
    for kv in fit["conditions"][0]["erev"]:
        region = kv["section"]
        for k, v in kv.items():
            if k == "section":
                continue
            ion = k[1:]
            ions.append((region, ion, float(v) * U.mV))

    return default, regs, ions, mechs, fit["fitting"][0]["junction_potential"]


def make_cell(swc, fit):
    morphology = A.load_swc_neuron(swc)

    dflt, regions, ions, mechanisms, offset = load_allen_fit(fit)

    labels = A.label_dict().add_swc_tags()
    labels["midpoint"] = "(location 0 0.5)"
    decor = A.decor()

    decor.set_property(
        tempK=dflt.temp,
        Vm=dflt.Vm,
        cm=dflt.cm,
        rL=dflt.rL,
    )

    for region, vs in regions:
        decor.paint(
            f'"{region}"',
            tempK=vs.temp,
            Vm=vs.Vm,
            cm=vs.cm,
            rL=vs.rL,
        )
    for region, ion, e in ions:
        decor.paint(f'"{region}"', ion=ion, rev_pot=e)
    decor.set_ion("ca", int_con=5e-5 * U.mM, ext_con=2.0 * U.mM, method="nernst/x=ca")

    for region, mech, values in mechanisms:
        nm = mech
        vs = {}
        sp = "/"
        for k, v in values.items():
            if mech == "pas" and k == "e":
                nm = f"{nm}{sp}{k}={v}"
                sp = ","
            else:
                vs[k] = v
        decor.paint(f'"{region}"', A.density(A.mechanism(nm, vs)))
    decor.place('"midpoint"', A.iclamp(0.2 * U.s, 1 * U.s, 150 * U.pA), "ic")
    decor.place('"midpoint"', A.threshold_detector(-40 * U.mV), "sd")
    
    return morphology, decor, labels

mrf, dec, lbl = make_cell("single_cell_allen.swc", "single_cell_allen_fit.json")
cvp = A.cv_policy_max_extent(20)

T = 1.4 * U.s
dt = 5 * U.us

reference = (
    1e3 * pl.read_csv("single_cell_allen_neuron_ref.csv")["U/mV"][:-1] - 14
)
```

```python
model = A.single_cell_model(A.cable_cell(mrf.morphology, dec, lbl, cvp))
model.probe("voltage", '"midpoint"', "Um", frequency=1 / (50 * U.us))
model.properties.catalogue.extend(A.allen_catalogue())
model.run(tfinal=T, dt=dt)

t0, U0 = model.traces[0].time, model.traces[0].value

fg, ax = plt.subplots()
ax.plot(t0, U0)
ax.plot(5e-3*np.arange(280000), reference)
ax.set_xlim(0, T.value_as(U.ms))
```

```python
def length(p, d):
    return np.sqrt((p.x - d.x)**2 + (p.y - d.y)**2 + (p.z - d.z)**2)

def lerp(a, b, f):
    return b*f + a*(1 - f)

def plerp(p, d, f):
    xp, yp, zp, rp = p.x, p.y, p.z, p.radius
    xd, yd, zd, rd = d.x, d.y, d.z, d.radius
    return A.mpoint(lerp(xp, xd, f), lerp(yp, yd, f), lerp(zp, zd, f), lerp(rp, rd, f))
```

# find fork

traverse tree from the leaves and the first fork point we can merge, s.t.

- both children originate from the same segment
- both children and the fork belong to the same region

```python
def find_fork(tree):
    terms = [n for n in range(tree.size) if tree.is_terminal(n)]
    forks = [n for n in range(tree.size) if tree.is_fork(n)]
    seen = {}
    for term in terms:
        seg = term
        tag = tree.segments[seg].tag
        hom = True
        segs = []
        while not tree.is_fork(seg):
            segs.append(seg)
            seg = tree.parents[seg]
            hom &= tag == tree.segments[seg].tag
        assert(tree.is_fork(seg))
        if segs and hom:
            if seg in seen and tree.segments[seen[seg][0]].tag == tag:
                break
            seen[seg] = segs
    else:
        raise RuntimeError("Couldn't find fork")
    tag = tree.segments[seg].tag
    for ix in segs:
        assert(tag == tree.segments[ix].tag)
    for ix in seen[seg]:
        assert(tag == tree.segments[ix].tag)
    return tag, seg, segs, seen[seg]
```

# prune tree; invariant lateral surface and length

replace fork with linear branch such that the total lateral area and length is invariant.
breaks invariance of the cross-sectional area

```python
def prune_tree_same_lateral(tree): 
    tag, c, ls, rs = find_fork(tree)

    # now split off the whole fork
    #     /              
    # --*      => --* + / + \
    #    \                 
    old, _ = tree.split_at(c)    
    
    # create a linear branch for replacement
    p = A.mnpos
    rep = A.segment_tree()

    # remember the last point
    x = tree.segments[c].dist.x
    y = tree.segments[c].dist.y
    z = tree.segments[c].dist.z

    # process indices into tuples
    ls = [(tree.segments[l].prox, tree.segments[l].dist) for l in ls]
    rs = [(tree.segments[r].prox, tree.segments[r].dist) for r in rs]

    # merge until one is exhausted
    while ls and rs:
        l_prox, l_dist = ls.pop()
        r_prox, r_dist = rs.pop()
        ll = length(l_prox, l_dist)
        lr = length(r_prox, r_dist)
        if ll == lr:
            p = rep.append(p, (x, y, z, l_prox.radius + r_prox.radius), (x + ll, y, z, l_dist.radius + r_dist.radius), tag=tag)
            x += ll
        elif ll < lr:
            # right is longer, so split it
            f = ll/lr
            split = plerp(r_prox, r_dist, f)
            # then append the length of l (shorter) to the tree
            p = rep.append(p, (x, y, z, l_prox.radius + r_prox.radius), (x + ll, y, z, l_dist.radius + split.radius), tag=tag)   
            # and the remaining bit goes into the todo-list
            rs.append((split, r_dist))
            x += ll
        elif ll > lr:
            # left is longer, so split it
            f = lr/ll
            split = plerp(l_prox, l_dist, f)
            # then append the length of l (shorter) to the tree
            p = rep.append(p, (x, y, z, l_prox.radius + r_prox.radius), (x + lr, y, z, r_dist.radius + split.radius), tag=tag)   
            # and the remaining bit goes into the todo-list
            ls.append((split, l_dist))
            x += ll
        else:
            raise RuntimeError("Impossible")
    
    # add on any remainders (only one something left)
    ls += rs
    while ls:
        prox, dist = ls.pop()
        p = rep.append(p, prox, dist, tag)
    
    return old.join_at(c, rep)
```

```python
new = mrf.segment_tree
ums = []
for i in range(10):
    new = prune_tree_same_lateral(new)
    model = A.single_cell_model(A.cable_cell(A.morphology(new), dec, lbl, cvp))
    model.probe("voltage", '"midpoint"', "Um", frequency=1 / (50 * U.us))
    model.properties.catalogue.extend(A.allen_catalogue())
    tbeg = pc()
    model.run(tfinal=T, dt=dt)
    tend = pc()
    print(f"iter={i:3d} time={tend - tbeg:.3f}")
    ums.append(model.traces[0].value)
```

```python
fg, ax = plt.subplots()
ax.plot(5e-3*np.arange(280000), reference, label='Reference, NRN')
ax.plot(t0, U0, label='Original')
for ix, um in enumerate(ums):
    ax.plot(t0, um, label=f'Pruned {ix + 1}')
ax.set_xlim(0, T.value_as(U.ms))
ax.legend()

fg.savefig('pruned.png')
```

```python
def prune_tree_same_cross(tree): 
    tag, c, ls, rs = find_fork(tree)

    # now split off the whole fork
    #     /              
    # --*      => --* + / + \
    #    \                 
    old, _ = tree.split_at(c)    
    
    # create a linear branch for replacement
    p = A.mnpos
    rep = A.segment_tree()

    # remember the last point
    x = tree.segments[c].dist.x
    y = tree.segments[c].dist.y
    z = tree.segments[c].dist.z

    # process indices into tuples
    ls = [(tree.segments[l].prox, tree.segments[l].dist) for l in ls]
    rs = [(tree.segments[r].prox, tree.segments[r].dist) for r in rs]

    # merge until one is exhausted
    while ls and rs:
        l_prox, l_dist = ls.pop()
        r_prox, r_dist = rs.pop()
        ll = length(l_prox, l_dist)
        lr = length(r_prox, r_dist)
        al = ll*(l_prox.radius + l_dist.radius)
        ar = lr*(r_prox.radius + r_dist.radius)
        rp = np.sqrt(l_prox.radius**2 + r_prox.radius**2)
        rd = np.sqrt(l_dist.radius**2 + r_dist.radius**2)
        ln = (al + ar)/(rp + rd)
        
        p = rep.append(p, (x, y, z, rp), (x + ln, y, z, rd), tag=tag)
        x += ln
    
    # add on any remainders (only one something left)
    ls += rs
    while ls:
        prox, dist = ls.pop()
        p = rep.append(p, prox, dist, tag)
    
    return old.join_at(c, rep)
```

```python
new = mrf.segment_tree
ums = []
for i in range(20):
    try:
        for _ in range(4):
            new = prune_tree_same_cross(new)
    except:
        break
    model = A.single_cell_model(A.cable_cell(A.morphology(new), dec, lbl, cvp))
    model.probe("voltage", '"midpoint"', "Um", frequency=1 / (50 * U.us))
    model.properties.catalogue.extend(A.allen_catalogue())
    tbeg = pc()
    model.run(tfinal=T, dt=dt)
    tend = pc()
    print(f"iter={i:3d} time={tend - tbeg:.3f}")
    ums.append(model.traces[0].value)
```

```python
def plot_morphology(tree, *, fg=None):
    colors = sns.color_palette('Set1')
    if not fg:
        fg = plt.figure(figsize=(10, 10))
    ax = fg.add_subplot(projection='3d')
    for seg in tree.segments:
        x0 = seg.prox.x
        x1 = seg.dist.x
        y0 = seg.prox.y
        y1 = seg.dist.y
        z0 = seg.prox.z
        z1 = seg.dist.z
        r0 = seg.prox.radius
        r1 = seg.dist.radius
        tag = seg.tag
        ax.plot(xs=[x0, x1], 
                ys=[y0, y1], 
                zs=[z0, z1], 
                color=colors[tag], 
                lw=(r0 + r1)) # average diameter from radius
    ax.set_xlabel(r'x $(\mu m)$')
    ax.set_ylabel(r'y $(\mu m)$')
    ax.set_zlabel(r'z $(\mu m)$')
    fg.tight_layout()
    return fg, ax
```
